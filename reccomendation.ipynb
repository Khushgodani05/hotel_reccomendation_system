```python
stop=stopwords.words("english")
filtered_text=[]
for i in text:
    filter=[]
    for words in i.split(" "):
        if words not in stop and words not in filter:
            filter.append(words)
    filtered_text.append(filter)
```

```python
from nltk.stem import WordNetLemmatizer
# nltk.download('averaged_perceptron_tagger_eng')
from nltk import pos_tag
lemmitizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
# nltk.download("wordnet")
def get_word_pos(treebank_tag):
    if treebank_tag.startswith("N"):
        return wordnet.NOUN
    elif treebank_tag.startswith("J"):
        return wordnet.ADJ
    elif treebank_tag.startswith("V"):
        return wordnet.VERB
    elif treebank_tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
lemma=[]
for i in part_of_speech:
    lemmed=[]
    for word,pos in i:
        lemmed.append(lemmitizer.lemmatize(word,get_word_pos(pos)))
    lemma.append(lemmed)

lemma[:5]
```

```python
final_tags=[]
for i in lemma:
    final_tags.append(" ".join(i))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
tags_processed=cv.fit_transform(final_tags).toarray()
```

```python
def preprocess(x):
    #re
    i=re.sub("[^a-zA-Z0-9]"," ",x)
    #tokenization
    i=i.strip().lower().split(" ")
    test=[]
    for words in i:
        if words!="":
            test.append(words)
    #stopword removal
    test1=[]
    for words in test:
        if words not in stop and words not in test1:
            test1.append(words)
    #lemmatization
    lemmed_sent=[]
    for word,pos in pos_tag(test1):
        lemmed_sent.append(lemmitizer.lemmatize(word,get_word_pos(pos)))
    #countvextorizer
    final_text=[]
    final_text.append(" ".join(lemmed_sent))
    vector=cv.transform(final_text).toarray()
    return vector
    
```

```python
def recemmond_hotel(country,description,data=data):
    description=preprocess(description)
    data1=data[data.country==country.lower()]
    tags=[preprocess(i) for i in data1.tags_combined]
    tags=np.array(tags).squeeze()
    cs= cosine_similarity(description,tags)
    data1["cosine_score"]=cs.reshape(-1,1)
    data1.sort_values(by="cosine_score",ascending=False,inplace=True)
    data1.sort_values(by="final_rating",ascending=False,inplace=True)
    data1.sort_values(by="review_no",ascending=False,inplace=True)
    # data1.sort_values(
    #     by=["cosine_score","final_rating","review_no"],
    #     ascending=[False,False,False],
    #     inplace=True,
    # )
    data1.drop_duplicates(subset="hotel_name",keep="first",inplace=True)
    data1.reset_index(inplace=True)
    return pd.DataFrame(data1.head(),columns=["hotel_name","hotel_address","final_rating"])
```
